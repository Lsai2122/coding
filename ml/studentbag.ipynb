```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
unfil_data = pd.read_csv("csvs/Noisy_Student_Bag_Price_Prediction_Dataset.csv")
```

```python
unfil_data
```

```python
for i in list(set(unfil_data["Size"])):
    print(f"{i} : \n{unfil_data[unfil_data["Size"]==i]["Price"].describe()}\n")
```

```python
obj = [col for col in unfil_data.columns if unfil_data[col].dtype == "object"]
oe_data = unfil_data.copy()
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
oe_data[obj] = oe.fit_transform(unfil_data[obj])
```

```python
#from sklearn.impute import KNNImputer
#imputer = KNNImputer(n_neighbors=3)
#filled_data = imputer.fit_transform(oe_data)
filled_data = pd.read_csv("csvs/Noisy_Student_Bag_Price_Prediction_Dataset_filled.csv")
```

```python
filled_data
```

```python
for i in obj:
    filled_data[i] = filled_data[i].astype("int64")
```

```python
filled_data
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe_data = filled_data.copy()
ohe = OneHotEncoder(sparse_output=False)
ohed = ohe.fit_transform(ohe_data[obj])
ohed = pd.DataFrame(ohed)
ohe_data = pd.concat([ohe_data, ohed], axis=1)
```

```python
ohe_data.drop(obj, axis=1, inplace=True)
```

```python
ohe_data.columns = ohe_data.columns.astype("str")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = ohe_data.copy()
scaled_data = scaler.fit_transform(scaled_data)
scaled_data = pd.DataFrame(scaled_data, columns=ohe_data.columns)
```

```python
scaled_data
```

```python
scaled_data["Price"]= ohe_data["Price"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
minmax_data = scaled_data.copy()
minmax_data = scaler.fit_transform(minmax_data)
minmax_data=pd.DataFrame(minmax_data)
```

```python
minmax_data
```

```python
x = minmax_data.drop("Price", axis=1)
y = minmax_data["Price"]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.metrics import mean_squared_error
trainx,testx,trainy,testy = train_test_split(x,y,test_size=0.2,random_state=0)
model = DecisionTreeRegressor(random_state=42)
param_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],  # Valid for regressors
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_features': [None, 'sqrt', 'log2'],
    'ccp_alpha': [0.0, 0.01, 0.1, 0.2]
}
# Randomized Search
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, 
                                   n_iter=50, cv=5, n_jobs=-1, scoring='accuracy', verbose=2, random_state=42)
random_search.fit(trainx,trainy)

# Best Parameters
print("Best Hyperparameters (Randomized Search):", random_search.best_params_)

```
